## Framing the problem

In this project, we will apply Machine Learning to predict the **weight (and BMI)** of people based on their eating habits and physical activities.

The data we'll be working with was garnered from a study conducted in Colombia, Peru and
Mexico. Participants were asked during a survey a multitude of questions regarding their eating habits, their
mode of transportation to work etc.

The data is present in CSV format. It's splitted in multiple files in the _data_ folder:
- X1 which is the features' training set
- Y1 which is the target's training set
- X2 which is the features' test set
- Y2 which is the target's test set

#### Values mapping
Compared to the survey some values have changed. We will map those values to the original answers given to the survey. It can be easily seen that there are two types of the categorical answers: ordinal and nominal ones. For the ordinal wee will put a number in front to preserve the right order of the responses.

```python
df['Consumption frequency of vegetables'] = df['Consumption frequency of vegetables'].replace(
    {1: '1. Never', 2: '2. Sometimes', 3: '3. Always'})
df['Number of main meals daily'] = df['Number of main meals daily'].replace(
    {1: '1', 2: '2', 3: '3', 4: '3+'})
df['Water consumption daily'] = df['Water consumption daily'].replace(
    {1: '1. Less than 1L', 2: '2. Between 1L and 2L', 3: '3. More than 2L'})
df['Physical activity frequency'] = df['Physical activity frequency'].replace(
    {0: '1. I do not have', 1: '2. 1 or 2 days', 2: '3. 2 or 4 days', 3: '4. 4 or 5 days'})
df['Time using technology devices'] = df['Time using technology devices'].replace(
    {0: '1. 0–2 hours', 1: '2. 3–5 hours', 2: '3. More than 5 hours'})
df['Food Consumption between meals'] = df['Food Consumption between meals'].replace(
    {'no' : '1. No', 'Sometimes' : '2. Sometimes', 'Frequently': '3. Frequently','Always': '4. Always'})
df['Alcohol consumption'] = df['Alcohol consumption'].replace(
    {'no' : '1. No', 'Sometimes' : '2. Sometimes', 'Frequently': '3. Frequently'})
    
```

### Create a Validation Set

We will put aside a validation set which will tell how well our future model performs. A good rule of thumb is to ensure that the test set is well representative of the various target data. To do this, we will create a **BMI** target as well as the different weight-classes so that the test set will be sampled according to these classes.

```python
df_split = df.copy()
```

```python
def get_classe(bmi):
    if 0 < bmi < 18.5:
        return 'underweight'
    if 18.5 < bmi < 25.0:
        return 'normal'
    if 25.0 < bmi < 30.0:
        return 'overweight'
    if bmi >=30.0:
        return 'obese'
```

```python
df_split['bmi'] = df_split.apply(lambda x : x['Weight']/x['Height']**2, axis=1)
df_split['class'] = df_split['bmi'].apply(lambda x : get_classe(x))
```

```python
df_split.head()
```

We can check the distribution of the classes.

```python
sns.countplot(df_split['class'], order=['underweight','normal','overweight','obese'])
```

We can already seen that there a big disproportionality between the classes. There is far more _normal_ classe than any other class.

```python
df_split['class'].value_counts()/len(df_split)*100
```

The _normal_ classe represent over 63% of the whole dataset.

Now as an exemple we will split the test and the training set with the sklearn's train_test_split method.

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df_split, test_size=0.2, random_state=42)
```

```python
test_set['class'].value_counts()/len(test_set)*100
```

The _obese_ and the _underweight_ classe in the test set now have the same counts which is not representative of the initial dataset.

As the test set is generated randomly and therefore to avoid a significant sampling bias, we will stratified the test set so that it will be as much representative of the overall data as possible.

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(df_split, df_split["class"]):
    strat_train_set = df_split.loc[train_index]
    strat_test_set = df_split.loc[test_index]
```

```python
strat_test_set['class'].value_counts()/len(strat_test_set)*100
```

We then drop the bmi and classe attributes and reset the indexes.

```python
strat_train_set.drop(['bmi','class'],axis=1, inplace=True)
strat_test_set.drop(['bmi','class'],axis=1, inplace=True)
```

```python
strat_train_set.reset_index(drop=True,inplace=True)
strat_test_set.reset_index(drop=True,inplace=True)
```

## Data preparation

In this section, we will change the data according to our analysis done previously.

### Feature encoding

```python
df_encode = df.copy()
```

#### Features binning

The feature _Usual transportation used_ contains much more samples the _Public Transport_ category than others. It can be useful to bin this feature in 2 categories rather than 5. For that we will change it in _Public Transport_ and all other categories will be combined in one category _Other_. Another reason to doing so is to avoid the curse of dimensionality. Indeed, if we would keep this feature as it is, we then would need to OneHotEncode it and we then would have 4 additional features.

```python
df_encode["Transportation"] = df_encode["Usual transportation used"].apply(lambda x : "Public_Transportation" if x == "Public_Transportation" else "Other")
```

```python
print_countplots(df_encode, ["Transportation"])
```

We then create a new feature _Public Transport_ which will represent a binary feature with values _yes_ or _no_ as if someone takes the public transportation or not.

```python
df_encode["Public Transportation"] = df_encode.apply(lambda x : "yes" if x['Transportation'] == "Public_Transportation" else "no", axis=1)
```

We drop the initial features

```python
df_encode.drop(['Usual transportation used','Transportation'], axis=1, inplace=True)
```

#### Encoding categorical features

The next step is to encode the categorical features. For the ordinal categorical features, we will representing it with a number from 1 to the number of feature's categories considered. For the binary nominal categorical features, we will encode it as 0 or 1. All other categories are OneHotEncode with the Panda's method get_dummies. (As we dropped the 'Usual Transportation Used' feature we don't have any)

```python
ordinal_cat = ['Consumption frequency of vegetables', 'Number of main meals daily', 'Food Consumption between meals',
               'Water consumption daily', 'Physical activity frequency', 'Time using technology devices',
               'Alcohol consumption']
for cat in ordinal_cat:
    df_encode[cat] = df_encode[cat].map(lambda x: sorted(df_encode[cat].unique()).index(x)+1)

# Handling binary categorical features
df_encode[['Family history with overweight', 'Consumption of high caloric food', 'Smoke', 'Calories consumption monitoring', 'Public Transportation']] = df_encode[[
    'Family history with overweight', 'Consumption of high caloric food', 'Smoke', 'Calories consumption monitoring', 'Public Transportation']].replace(to_replace=['no', 'yes'], value=[0, 1])
df_encode[['Gender']] = df_encode[['Gender']].replace(
    to_replace=['Female', 'Male'], value=[0, 1])
```

```python
df_encode.head()
```

### Features selection

#### Looking at the correlation

To check if any feature is redundant we will use the linear correlation to detect if we should drop some of the features.

For better clarity , we will cut away half of the correlation matrix along the diagonal line and keep only the lower triangle.

```python
plt.figure(figsize=(15, 6))
mask = np.triu(np.ones_like(df_encode.corr(), dtype=np.bool))
heatmap = sns.heatmap(df_encode.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

#### Looking at the variance

Another method to select which features to keep, is checking at the variance of each feature. As a feature has a low variance it doesn't add much new information to the dataset and can be dropped. We choose 0.2 as a threshold.

```python
df_encode.var()
```

The features 'Smoke' and 'Calories consumption monitoring' variances are situated below our threshold and will be dropped.

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(0.2)
selector.fit_transform(df_encode)
selected_features = selector.get_feature_names_out()
selected_features
```

```python
df_encode = df_encode[selected_features]
```

```python
df_encode.head()
```

### Age feature transformation

The last feature engineering is to the _Age_ feature. In the visualization section, we saw that the distribution is skewed to the left.

```python
df_encode['Age'].skew()
```

We will apply the log function to minimize this skewness and also remove the outliers present.

```python
df_encode['Age'] = df_encode['Age'].map(lambda x: np.log(x))
```

```python
df_encode['Age'].skew()
```

We can see that this transformation improves the linear correlation to the weight.

```python
df_encode.corr()['Weight'].sort_values(0, False)[1:]['Age']
```

### Handling outliers

For the Age attribute and the Height attribute we will clip the values that exceed the 95th percentile and are lower than the 5th percentile. We do this instead of discarding such outliers because we can expect that our model will encounter the same outliers in the validation set as well as in the dataset X2 prediction.

```python
for col in ['Height','Age']:
    upper_lim = df_encode[col].quantile(.95)
    lower_lim = df_encode[col].quantile(.05)

    df_encode.loc[(df[col] > upper_lim), col] = upper_lim
    df_encode.loc[(df[col] < lower_lim), col] = lower_lim
```

```python
df_encode.corr()['Weight'].sort_values(0, False)[1:]
```

### Data scaling

We will now proceed to scaling our data. This part is fundamental. If no scaling is operated, a little variation is one feature can have huge repercution compared to other features.

```python
df_scaled = df_encode.copy()
```

### Adding up all the transformations

```python
from sklearn.base import BaseEstimator
class DataEngineering(BaseEstimator):
    def __init__(self): # no *args or **kargs
        pass
    def fit(self, X, y=None):
        return self # nothing else to do
    def transform(self, X, y=None, augmented=True):
        columns = ['Gender', 'Age', 'Height', 'Family history with overweight', 'Consumption of high caloric food',
           'Consumption frequency of vegetables', 'Number of main meals daily', 'Food Consumption between meals', 'Smoke',
           'Water consumption daily', 'Calories consumption monitoring', 'Physical activity frequency', 'Time using technology devices',
           'Alcohol consumption', 'Usual transportation used']
        X.columns = columns
        X['Height'] = X['Height']*100
        X['Consumption frequency of vegetables'] = X['Consumption frequency of vegetables'].replace(
            {1: '1. Never', 2: '2. Sometimes', 3: '3. Always'})
        X['Number of main meals daily'] = X['Number of main meals daily'].replace(
            {1: '1', 2: '2', 3: '3', 4: '3+'})
        X['Water consumption daily'] = X['Water consumption daily'].replace(
            {1: '1. Less than 1L', 2: '2. Between 1L and 2L', 3: '3. More than 2L'})
        X['Physical activity frequency'] = X['Physical activity frequency'].replace(
            {0: '1. I do not have', 1: '2. 1 or 2 days', 2: '3. 2 or 4 days', 3: '4. 4 or 5 days'})
        X['Time using technology devices'] = X['Time using technology devices'].replace(
            {0: '1. 0–2 hours', 1: '2. 3–5 hours', 2: '3. More than 5 hours'})
        X['Food Consumption between meals'] = X['Food Consumption between meals'].replace(
            {'no' : '1. No', 'Sometimes' : '2. Sometimes', 'Frequently': '3. Frequently','Always': '4. Always'})
        X['Alcohol consumption'] = X['Alcohol consumption'].replace(
            {'no' : '1. No', 'Sometimes' : '2. Sometimes', 'Frequently': '3. Frequently'})

        X["Transportation"] = X["Usual transportation used"].apply(lambda x : "Public_Transportation" if x == "Public_Transportation" else "Other")
        X["Public Transportation"] = X.apply(lambda x : "yes" if x['Transportation'] == "Public_Transportation" else "no", axis=1)
        X.drop(["Usual transportation used",
               "Transportation",'Smoke','Calories consumption monitoring'
               ], inplace=True, axis=1,errors='ignore')
               
        ordinal_cat = ['Consumption frequency of vegetables', 'Number of main meals daily', 'Food Consumption between meals',
               'Water consumption daily', 'Physical activity frequency', 'Time using technology devices',
               'Alcohol consumption']
        for cat in ordinal_cat:
            X[cat] = X[cat].map(lambda x: sorted(X[cat].unique()).index(x)+1)

        # Handling binary categorical features
        X[['Family history with overweight', 'Consumption of high caloric food','Public Transportation']] = X[[
            'Family history with overweight', 'Consumption of high caloric food','Public Transportation']].replace(to_replace=['no', 'yes'], value=[0, 1])
        # Handling binary categorical features
        X['Gender'] = X['Gender'].replace(to_replace=['Male', 'Female'], value=[0, 1])
        selector = VarianceThreshold(0.2)
        selector.fit_transform(X,y)
        selected_features = selector.get_feature_names_out()
        X = X[selected_features]
        
        if augmented:
            df_augmented = pd.concat([X,y],axis=1)
            df_smogn = smogn.smoter(
                ## main arguments
                data = df_augmented,           ## pandas dataframe
                y = 'Weight',          ## string ('header name')
                k = 7,
                pert = 0.02,                  ## positive integer (k < n)
                samp_method = 'extreme',  ## string ('balance' or 'extreme')

                # ## phi relevance arguments
                rel_thres = 0.6,         ## positive real number (0 < R < 1)
                rel_method = 'auto',      ## string ('auto' or 'manual')
                rel_xtrm_type = 'both',   ## string ('low' or 'both' or 'high')
                rel_coef = 0.4          ## positive real number (0 < R)
            )
            X = pd.get_dummies(df_smogn.drop('Weight',axis=1))
            y = pd.DataFrame(df_smogn['Weight'],columns=['Weight'])
        
        for col in ['Height','Age']:
            upper_lim = X[col].quantile(.95)
            lower_lim = X[col].quantile(.05)

            X.loc[(X[col] > upper_lim), col] = upper_lim
            X.loc[(X[col] < lower_lim), col] = lower_lim
            
        X['Age'] = X['Age'].map(lambda x: np.log(x))
        
        if y is not None:
            col = y.columns[0]
            upper_lim = y[col].quantile(.95)
            lower_lim = y[col].quantile(.05)

            y.loc[(y[col] > upper_lim), col] = upper_lim
            y.loc[(y[col] < lower_lim), col] = lower_lim
            
            return X,y['Weight']

        return X
```

### Evaluation function

We will use prediction criterion that will be balanced according to the different weight-classes.

```python
def score_weight_class(bmi_pred, bmi_true, low, high):
    tol = 1
    vpred = (bmi_pred >= low - tol) & (bmi_pred < high+tol)
    vtrue = (bmi_true >= low) & (bmi_true < high)
    if vtrue.sum() == 0:
        print("no true samples here")
        return 0
    rmse = np.sqrt(((bmi_true[vtrue]-bmi_pred[vtrue])**2).mean())
    rmse = rmse/(high-low+tol)  # normalize rmse in interval
    acc = (vpred & vtrue).sum()/vtrue.sum()
    return rmse*(1-acc)


def score_regression_weight(ytrue, ypred, height):
    bmi_pred = ypred/(height*height)
    bmi_true = ytrue/(height*height)
    scores = []
    for bmi_low, bmi_high in zip([0, 18.5, 25, 30], [18.5, 25, 30, 100]):
        scores.append(score_weight_class(bmi_pred, bmi_true,
                                         low=bmi_low, high=bmi_high))
    return np.mean(scores)
```

#### Running RandomizedSearchCV with mlflow

We will be using [Mlflow](https://mlflow.org/) for logging parameters and metrics when running the Machine Learning models. This will allow us to preserve all the performance of our models as well as visualize the results. 

(To run the UI , run _mlflow ui_ in the terminal in the folder of this project.)

```python
import mlflow
import mlflow.sklearn
mlflow.set_experiment("Obsesity Regression")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer,TransformedTargetRegressor
from sklearn.preprocessing import MinMaxScaler,PowerTransformer, QuantileTransformer

num_transformer_standard =ColumnTransformer(
    transformers=[('Scaler', StandardScaler(),attribs)], remainder='passthrough'
)

num_transformer_min_max =ColumnTransformer(
    transformers=[('Scaler', MinMaxScaler(),attribs)], remainder='passthrough'
)

def run_randomized_search(X_train,y_train,X_test,y_test,models,params,n_iter=200, augmented=True):
    for key in models.keys():
        with mlflow.start_run():
            print(f"Running RandomizedSearchCV for {key}.")
            model = models[key]
            param = {f'regressor__{k}' : v for k,v in params.items() if key in k or 'PCA' in k}
            pipeline = Pipeline(steps=[
                ('NumTrans', num_transformer_min_max if key=='MLPRegressor' else num_transformer_standard),
                ('PCA', PCA()),
                (key,model)
            ])
            # We tranform the target to normalize it as much as possible 
            trans = TransformedTargetRegressor(regressor=pipeline, transformer=PowerTransformer())

            mlflow.log_param("Best_Estimator", key)
            
            rs = RandomizedSearchCV(trans, param, n_iter=n_iter, cv=KFold(n_splits=5), n_jobs=-2,
                                    verbose=10, scoring='neg_root_mean_squared_error', refit='neg_root_mean_squared_error',
                                    random_state=42
                                    )
            
            rs.fit(X_train, y_train)
            
            mlflow.log_metric("score_train_rmse", abs(rs.best_score_))
            preds = rs.predict(X_test)
            preds_val = rs.predict(X_validation)
            mlflow.log_metric("score_test_rmse", mean_squared_error(y_test,preds, squared=False))
            mlflow.log_metric("score_val_rmse", mean_squared_error(y_validation,preds_val, squared=False))
        
            mlflow.log_metric("score_test_reg", score_regression_weight(y_test.ravel(),preds.ravel(),X_test['Height'].ravel()/100))
            mlflow.log_metric("score_val_reg", score_regression_weight(y_validation.ravel(),preds_val.ravel(),X_validation['Height'].ravel()/100))
            if augmented:
                mlflow.log_param("Augmentation", "YES")
            else:
                mlflow.log_param("Augmentation", "NO")
        
            for key in rs.best_params_:
                mlflow.log_param(key.split('__')[2], rs.best_params_[key])
            mlflow.sklearn.log_model(rs.best_estimator_, "best_model")
```

#### Adding a additional non linear estimator

The Voting ensemble estimator will take the average of all the estimators it contains. In that way we may improve the general perforamnce.

```python
def get_best_model_run_id(name, scoring='score_val_reg', augmented=False):
    if augmented:
        df_experients = mlflow.search_runs(filter_string=f"params.Best_Estimator='{name}' and params.Augmentation='YES'")
    else :
        df_experients = mlflow.search_runs(filter_string=f"params.Best_Estimator='{name}' and params.Augmentation='NO'")
    #Getting the best run id
    return df_experients.loc[df_experients[f'metrics.{scoring}'].idxmin()]['run_id']
```

```python
from sklearn.ensemble import VotingRegressor

def run_voting_regressor(X_train,y_train,X_test,y_test, augmented=False):
    #Getting the best experients from Mlflow
    ridge = mlflow.sklearn.load_model("runs:/" + get_best_model_run_id('Ridge',augmented=augmented) + "/best_model")
    lasso = mlflow.sklearn.load_model("runs:/" + get_best_model_run_id('Lasso',augmented=augmented) + "/best_model")
    elastic = mlflow.sklearn.load_model("runs:/" + get_best_model_run_id('ElasticNet',augmented=augmented) + "/best_model")
    
    with mlflow.start_run():
        vr = VotingRegressor([('ridge',ridge),('lasso',lasso),('elastic',elastic)])
        vr.fit(X_train, y_train)

        mlflow.log_param("Best_Estimator", 'Voting')
        preds = vr.predict(X_test)
        preds_val = vr.predict(X_validation)
        mlflow.log_metric("score_test_rmse", mean_squared_error(y_test,preds, squared=False))
        mlflow.log_metric("score_val_rmse", mean_squared_error(y_validation,preds_val, squared=False))

        mlflow.log_metric("score_test_reg", score_regression_weight(y_test.ravel(),preds.ravel(),X_test['Height'].ravel()/100))
        mlflow.log_metric("score_val_reg", score_regression_weight(y_validation.ravel(),preds_val.ravel(),X_validation['Height'].ravel()/100))
        if augmented:
            mlflow.log_param("Augmentation", "YES")
        else:
            mlflow.log_param("Augmentation", "NO")
              
```
